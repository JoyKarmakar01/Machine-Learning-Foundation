#### 1. Regression Algorithms

- **Ridge Regression**
- **Lasso Regression**
- **Elastic Net Regression**
- **Polynomial Regression**
- **Support Vector Regression (SVR)**
- **Decision Tree Regression**
- **Random Forest Regression**
- **Gradient Boosting Regression**
- **AdaBoost Regression**
- **XGBoost Regression**
- **LightGBM Regression**
- **CatBoost Regression**

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings("ignore")

# Load dataset (using Boston housing dataset as an example)
def load_and_prepare_data():
    cali_housing = fetch_california_housing()
    X = cali_housing.data
    y = cali_housing.target

    # Scale the data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Split data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
    return X_train, X_test, y_train, y_test

# Evaluate and print the performance of regressors
def evaluate_regressor(name, regressor, X_train, X_test, y_train, y_test):
    regressor.fit(X_train, y_train)
    y_pred = regressor.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"--- {name} ---")
    print(f"Mean Squared Error: {mse:.4f}\n")

# Define all regression models
def get_regressors():
    return [
        ("Linear Regression", LinearRegression()),
        ("Ridge Regression", Ridge(alpha=1.0)),
        ("Lasso Regression", Lasso(alpha=0.1)),
        ("Elastic Net Regression", ElasticNet(alpha=0.1, l1_ratio=0.5)),
        ("Support Vector Regression (SVR)", SVR(kernel='rbf', C=1.0, epsilon=0.1)),
        ("Decision Tree Regression", DecisionTreeRegressor(random_state=42)),
        ("Random Forest Regression", RandomForestRegressor(n_estimators=100, random_state=42)),
        ("Gradient Boosting Regression", GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)),
        ("AdaBoost Regression", AdaBoostRegressor(n_estimators=100, random_state=42)),
        ("XGBoost Regression", xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)),
        ("LightGBM Regression", lgb.LGBMRegressor(n_estimators=100, learning_rate=0.1, random_state=42, force_col_wise=True)),
        ("CatBoost Regression", CatBoostRegressor(n_estimators=100, learning_rate=0.1, depth=3, random_state=42, verbose=0))
    ]

# Polynomial Regression (degree 2) function
def polynomial_regression(X_train, X_test, y_train, y_test):
    poly = PolynomialFeatures(degree=2)
    X_poly_train = poly.fit_transform(X_train)
    X_poly_test = poly.transform(X_test)
    poly_regressor = LinearRegression()
    evaluate_regressor("Polynomial Regression (degree 2)", poly_regressor, X_poly_train, X_poly_test, y_train, y_test)

# Main function to run all regressors
def main():
    X_train, X_test, y_train, y_test = load_and_prepare_data()
    
    # Evaluate non-polynomial regressors
    regressors = get_regressors()
    for name, regressor in regressors:
        evaluate_regressor(name, regressor, X_train, X_test, y_train, y_test)
    
    # Evaluate polynomial regression
    polynomial_regression(X_train, X_test, y_train, y_test)

if __name__ == "__main__":
    main()
```

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis, LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, classification_report

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings("ignore")

# Load dataset (Iris dataset for classification)
def load_and_prepare_data():
    iris = load_iris()
    X = iris.data
    y = iris.target
    
    # Scale the data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Split data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
    return X_train, X_test, y_train, y_test

# Evaluate and print the performance of classifiers
def evaluate_classifier(name, classifier, X_train, X_test, y_train, y_test):
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"--- {name} ---")
    print(f"Accuracy: {accuracy:.4f}")
    print(classification_report(y_test, y_pred))
    print("\n")

# Define all classification models
def get_classifiers():
    return [
        ("Logistic Regression", LogisticRegression(max_iter=200)),
        ("K-Nearest Neighbors (KNN)", KNeighborsClassifier(n_neighbors=5)),
        ("Support Vector Machine (SVM)", SVC(kernel='linear')),
        ("Decision Tree Classification", DecisionTreeClassifier(random_state=42)),
        ("Random Forest Classification", RandomForestClassifier(n_estimators=100, random_state=42)),
        ("Gradient Boosting Classification", GradientBoostingClassifier(n_estimators=100, random_state=42)),
        ("AdaBoost Classification", AdaBoostClassifier(n_estimators=100, random_state=42)),
        ("XGBoost Classification", xgb.XGBClassifier(n_estimators=100, learning_rate=0.1, random_state=42)),
        ("LightGBM Classification", lgb.LGBMClassifier(n_estimators=100, learning_rate=0.1, random_state=42)),
        ("CatBoost Classification", CatBoostClassifier(n_estimators=100, learning_rate=0.1, depth=3, random_state=42, verbose=0)),
        ("Gaussian Naive Bayes", GaussianNB()),
       #("Multinomial Naive Bayes", MultinomialNB()),
        ("Bernoulli Naive Bayes", BernoulliNB()),
        ("Quadratic Discriminant Analysis (QDA)", QuadraticDiscriminantAnalysis()),
        ("Linear Discriminant Analysis (LDA)", LinearDiscriminantAnalysis())
    ]

# Main function to run all classifiers
def main():
    X_train, X_test, y_train, y_test = load_and_prepare_data()
    
    classifiers = get_classifiers()
    for name, classifier in classifiers:
        evaluate_classifier(name, classifier, X_train, X_test, y_train, y_test)

if __name__ == "__main__":
    main()
```

### Unsupervised Learning

#### 3. Clustering Algorithms
- **K-Means Clustering**
- **K-Medoids Clustering**
- **Hierarchical Clustering (Agglomerative, Divisive)**
- **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**
- **OPTICS (Ordering Points To Identify the Clustering Structure)**
- **Mean Shift Clustering**
- **Gaussian Mixture Models (GMM)**
- **BIRCH (Balanced Iterative Reducing and Clustering using Hierarchies)**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_wine, make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, MeanShift, Birch
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn.cluster import OPTICS
from sklearn_extra.cluster import KMedoids
from scipy.cluster.hierarchy import linkage, fcluster

def load_data_for_algorithms():
    """Load datasets suitable for each clustering algorithm."""
    data = {}
    
    # Load Iris dataset for K-Means and GMM
    iris = load_iris()
    data['K-Means Clustering'] = iris.data
    
    # Synthetic dataset for K-Medoids and DBSCAN
    data['K-Medoids Clustering'] = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)[0]
    data['DBSCAN'] = data['K-Medoids Clustering']
    
    # Wine dataset for Hierarchical Clustering
    wine = load_wine()
    data['Hierarchical Clustering'] = wine.data
    
    # Synthetic dataset for OPTICS
    data['OPTICS'] = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)[0]
    
    # Blobs dataset for Mean Shift
    blobs = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
    data['Mean Shift Clustering'] = blobs[0]
    
    # Custom synthetic dataset for BIRCH
    data['BIRCH'] = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)[0]
    
    return data

def plot_clustering(X, labels, title):
    """Plot clustering results."""
    try:
        plt.figure(figsize=(8, 6))
        plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
        plt.title(title)
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.colorbar()
        plt.show()
    except Exception as e:
        print(f"Error plotting clustering results: {e}")

def evaluate_clustering(name, X, labels):
    """Evaluate and plot clustering results."""
    try:
        if len(set(labels)) > 1:
            silhouette_avg = silhouette_score(X, labels)
            print(f"--- {name} ---")
            print(f"Silhouette Score: {silhouette_avg:.4f}")
        else:
            print(f"--- {name} ---")
            print("Single cluster detected, silhouette score not computed.")
        plot_clustering(X, labels, name)
    except Exception as e:
        print(f"Error evaluating clustering: {e}")

def run_clustering_algorithms(data):
    """Run and evaluate various clustering algorithms on different datasets."""
    try:
        algorithms = [
            ("K-Means Clustering", KMeans(n_clusters=3, random_state=42)),
            ("K-Medoids Clustering", KMedoids(n_clusters=3, random_state=42)),
            ("Hierarchical Clustering - Agglomerative", 'Agglomerative'),
            ("Hierarchical Clustering - Divisive", 'Divisive'),
            ("DBSCAN", DBSCAN(eps=0.5, min_samples=5)),
            ("OPTICS", OPTICS(min_samples=5)),
            ("Mean Shift Clustering", MeanShift()),
            ("Gaussian Mixture Models (GMM)", GaussianMixture(n_components=3, random_state=42)),
            ("BIRCH", Birch(n_clusters=3))
        ]

        for name, algorithm in algorithms:
            try:
                X = data[name]
                if name == "Hierarchical Clustering - Agglomerative":
                    Z = linkage(X, 'ward')
                    labels = fcluster(Z, 3, criterion='maxclust')
                elif name == "Hierarchical Clustering - Divisive":
                    # Divisive clustering is not directly available in scikit-learn.
                    # Here we only show Agglomerative for simplicity.
                    Z = linkage(X, 'ward')
                    labels = fcluster(Z, 3, criterion='maxclust')
                else:
                    model = algorithm.fit(X)
                    if name == "Gaussian Mixture Models (GMM)":
                        labels = model.predict(X)
                    else:
                        labels = model.labels_
                evaluate_clustering(name, X, labels)
            except Exception as e:
                print(f"Error running {name}: {e}")
    except Exception as e:
        print(f"Error running clustering algorithms: {e}")

def main():
    """Main function to load data and run clustering algorithms."""
    data = load_data_for_algorithms()
    run_clustering_algorithms(data)

if __name__ == "__main__":
    main()
```

### Unsupervised Learning


#### 4. Dimensionality Reduction Algorithms
- **Principal Component Analysis (PCA)**
- **Linear Discriminant Analysis (LDA)**
- **Kernel PCA**
- **t-Distributed Stochastic Neighbor Embedding (t-SNE)**
- **Independent Component Analysis (ICA)**
- **Factor Analysis**



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA, KernelPCA, FastICA, FactorAnalysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

def load_and_prepare_data():
    """
    Load the Iris dataset and prepare it for dimensionality reduction.
    """
    # Load the Iris dataset
    iris = load_iris()
    X = iris.data
    y = iris.target

    # Standardize the features (important for many dimensionality reduction algorithms)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    return X_scaled, y

def apply_pca(X):
    """
    Apply Principal Component Analysis (PCA) to reduce the dimensions of the data.

    Parameters:
    X (array-like): The input data.

    Returns:
    pca_result (array-like): The data after applying PCA.
    """
    pca = PCA(n_components=2)  # Reduce to 2 components for visualization
    pca_result = pca.fit_transform(X)

    return pca_result

def apply_lda(X, y):
    """
    Apply Linear Discriminant Analysis (LDA) to reduce the dimensions of the data.

    Parameters:
    X (array-like): The input data.
    y (array-like): The target labels.

    Returns:
    lda_result (array-like): The data after applying LDA.
    """
    lda = LinearDiscriminantAnalysis(n_components=2)  # Reduce to 2 components for visualization
    lda_result = lda.fit_transform(X, y)

    return lda_result

def apply_kernel_pca(X):
    """
    Apply Kernel PCA to reduce the dimensions of the data using a kernel method.

    Parameters:
    X (array-like): The input data.

    Returns:
    kernel_pca_result (array-like): The data after applying Kernel PCA.
    """
    kernel_pca = KernelPCA(n_components=2, kernel='rbf')  # Radial basis function kernel
    kernel_pca_result = kernel_pca.fit_transform(X)

    return kernel_pca_result

def apply_tsne(X):
    """
    Apply t-Distributed Stochastic Neighbor Embedding (t-SNE) for dimensionality reduction.

    Parameters:
    X (array-like): The input data.

    Returns:
    tsne_result (array-like): The data after applying t-SNE.
    """
    tsne = TSNE(n_components=2, random_state=42)  # Reduce to 2 components
    tsne_result = tsne.fit_transform(X)

    return tsne_result

def apply_ica(X):
    """
    Apply Independent Component Analysis (ICA) to separate mixed signals.

    Parameters:
    X (array-like): The input data.

    Returns:
    ica_result (array-like): The data after applying ICA.
    """
    ica = FastICA(n_components=2, random_state=42)  # Reduce to 2 components
    ica_result = ica.fit_transform(X)

    return ica_result

def apply_factor_analysis(X):
    """
    Apply Factor Analysis to reduce the dimensions of the data.

    Parameters:
    X (array-like): The input data.

    Returns:
    factor_analysis_result (array-like): The data after applying Factor Analysis.
    """
    factor_analysis = FactorAnalysis(n_components=2)  # Reduce to 2 components
    factor_analysis_result = factor_analysis.fit_transform(X)

    return factor_analysis_result

def plot_results(results, labels, title):
    """
    Plot the results of dimensionality reduction.

    Parameters:
    results (array-like): The results of the dimensionality reduction.
    labels (array-like): The target labels.
    title (str): The title of the plot.
    """
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(results[:, 0], results[:, 1], c=labels, cmap='viridis')
    plt.colorbar(scatter, label='Class Label')
    plt.title(title)
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.show()

def main():
    """
    Main function to apply all dimensionality reduction algorithms and plot results.
    """
    # Load and prepare data
    X, y = load_and_prepare_data()

    # Apply PCA and plot results
    pca_result = apply_pca(X)
    plot_results(pca_result, y, 'PCA Result')

    # Apply LDA and plot results
    lda_result = apply_lda(X, y)
    plot_results(lda_result, y, 'LDA Result')

    # Apply Kernel PCA and plot results
    kernel_pca_result = apply_kernel_pca(X)
    plot_results(kernel_pca_result, y, 'Kernel PCA Result')

    # Apply t-SNE and plot results
    tsne_result = apply_tsne(X)
    plot_results(tsne_result, y, 't-SNE Result')

    # Apply ICA and plot results
    ica_result = apply_ica(X)
    plot_results(ica_result, y, 'ICA Result')

    # Apply Factor Analysis and plot results
    factor_analysis_result = apply_factor_analysis(X)
    plot_results(factor_analysis_result, y, 'Factor Analysis Result')

if __name__ == "__main__":
    main()
```

### Other Important Techniques

#### 5. Ensemble Methods
- **Bagging (Bootstrap Aggregating)**
- **Boosting (AdaBoost, Gradient Boosting, XGBoost, LightGBM, CatBoost)**
- **Stacking (Stacked Generalization)**
- **Voting Classifier**


```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import (
    BaggingClassifier,
    AdaBoostClassifier,
    GradientBoostingClassifier,
    VotingClassifier,
    StackingClassifier
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier

def load_and_prepare_data():
    """
    Load the Iris dataset and prepare it for classification.
    """
    iris = load_iris()
    X = iris.data
    y = iris.target
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    return X_train, X_test, y_train, y_test

def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    """
    Train and evaluate a model, and print the accuracy.

    Parameters:
    model: The machine learning model to be evaluated.
    X_train (array-like): Training data features.
    X_test (array-like): Testing data features.
    y_train (array-like): Training data labels.
    y_test (array-like): Testing data labels.
    model_name (str): Name of the model for display.
    """
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy of {model_name}: {accuracy:.4f}")

def main():
    """
    Main function to demonstrate various ensemble methods.
    """
    # Load and prepare data
    X_train, X_test, y_train, y_test = load_and_prepare_data()

    # Bagging
    print("--- Bagging ---")
    bagging_model = BaggingClassifier(n_estimators=50, random_state=42)
    evaluate_model(bagging_model, X_train, X_test, y_train, y_test, "Bagging")

    # Boosting
    print("\n--- Boosting Algorithms ---")

    # AdaBoost
    adaboost_model = AdaBoostClassifier(n_estimators=50, random_state=42)
    evaluate_model(adaboost_model, X_train, X_test, y_train, y_test, "AdaBoost")

    # Gradient Boosting
    gradient_boosting_model = GradientBoostingClassifier(n_estimators=50, random_state=42)
    evaluate_model(gradient_boosting_model, X_train, X_test, y_train, y_test, "Gradient Boosting")

    # XGBoost
    xgboost_model = xgb.XGBClassifier(n_estimators=50, use_label_encoder=False, eval_metric='mlogloss', random_state=42)
    evaluate_model(xgboost_model, X_train, X_test, y_train, y_test, "XGBoost")

    # LightGBM
    lightgbm_model = lgb.LGBMClassifier(n_estimators=50, random_state=42)
    evaluate_model(lightgbm_model, X_train, X_test, y_train, y_test, "LightGBM")

    # CatBoost
    catboost_model = CatBoostClassifier(n_estimators=50, verbose=0, random_state=42)
    evaluate_model(catboost_model, X_train, X_test, y_train, y_test, "CatBoost")

    # Stacking
    print("\n--- Stacking ---")
    estimators = [
        ('decision_tree', DecisionTreeClassifier()),
        ('knn', KNeighborsClassifier())
    ]
    stacking_model = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
    evaluate_model(stacking_model, X_train, X_test, y_train, y_test, "Stacking")

    # Voting Classifier
    print("\n--- Voting Classifier ---")
    voting_model = VotingClassifier(estimators=[
        ('decision_tree', DecisionTreeClassifier()),
        ('knn', KNeighborsClassifier()),
        ('svc', SVC(probability=True))
    ], voting='soft')
    evaluate_model(voting_model, X_train, X_test, y_train, y_test, "Voting Classifier")

if __name__ == "__main__":
    main()
```

#### 6. Neural Networks and Deep Learning
- **Artificial Neural Networks (ANN)**
- **Convolutional Neural Networks (CNN)**
- **Recurrent Neural Networks (RNN)**
- **Long Short-Term Memory Networks (LSTM)**
- **Gated Recurrent Units (GRU)**
- **Autoencoders**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, SimpleRNN, LSTM, GRU
from tensorflow.keras.datasets import mnist
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

# Load and prepare the dataset
def load_and_prepare_data():
    # Load MNIST dataset
    (X_train, y_train), (X_test, y_test) = mnist.load_data()
    
    # Preprocess MNIST data
    X_train = X_train.reshape((X_train.shape[0], 28, 28, 1)) / 255.0
    X_test = X_test.reshape((X_test.shape[0], 28, 28, 1)) / 255.0
    
    # Preprocess labels
    y_train = tf.keras.utils.to_categorical(y_train, 10)
    y_test = tf.keras.utils.to_categorical(y_test, 10)
    
    return (X_train, y_train), (X_test, y_test)

# Create a simple ANN model
def build_ann_model(input_shape, num_classes):
    model = Sequential([
        Flatten(input_shape=input_shape),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Create a simple CNN model
def build_cnn_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        Flatten(),
        Dense(64, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# # Create a simple RNN model
# def build_rnn_model(input_shape, num_classes):
#     model = Sequential([
#         SimpleRNN(50, input_shape=input_shape, activation='relu'),
#         Dense(num_classes, activation='softmax')
#     ])
#     model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
#     return model

# # Create an LSTM model
# def build_lstm_model(input_shape, num_classes):
#     model = Sequential([
#         LSTM(50, input_shape=input_shape, activation='relu'),
#         Dense(num_classes, activation='softmax')
#     ])
#     model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
#     return model

# # Create a GRU model
# def build_gru_model(input_shape, num_classes):
#     model = Sequential([
#         GRU(50, input_shape=input_shape, activation='relu'),
#         Dense(num_classes, activation='softmax')
#     ])
#     model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
#     return model

# Main function to execute all models
def main():
    (X_train, y_train), (X_test, y_test) = load_and_prepare_data()
    
    # ANN
    print("--- Artificial Neural Networks (ANN) ---")
    ann_model = build_ann_model((28, 28, 1), 10)
    ann_model.fit(X_train, y_train, epochs=1, batch_size=64, validation_split=0.2)
    
    # CNN
    print("--- Convolutional Neural Networks (CNN) ---")
    cnn_model = build_cnn_model((28, 28, 1), 10)
    cnn_model.fit(X_train, y_train, epochs=1, batch_size=64, validation_split=0.2)
    
#     # RNN
#     print("--- Recurrent Neural Networks (RNN) ---")
#     # Reshape data for RNN
#     X_train_rnn = np.expand_dims(X_train, axis=1)
#     X_test_rnn = np.expand_dims(X_test, axis=1)
#     rnn_model = build_rnn_model((1, 28, 28), 10)
#     rnn_model.fit(X_train_rnn, y_train, epochs=1, batch_size=64, validation_split=0.2)
    
#     # LSTM
#     print("--- Long Short-Term Memory Networks (LSTM) ---")
#     lstm_model = build_lstm_model((1, 28, 28), 10)
#     lstm_model.fit(X_train_rnn, y_train, epochs=1, batch_size=64, validation_split=0.2)
    
#     # GRU
#     print("--- Gated Recurrent Units (GRU) ---")
#     gru_model = build_gru_model((1, 28, 28), 10)
#     gru_model.fit(X_train_rnn, y_train, epochs=1, batch_size=64, validation_split=0.2)

if __name__ == "__main__":
    main()
```

#### 7. Anomaly Detection
- **Isolation Forest**
- **Local Outlier Factor (LOF)**
- **One-Class SVM**






```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

def generate_synthetic_data():
    """
    Generate synthetic data for anomaly detection.
    """
    # Create a dataset with two clusters and some outliers
    X, _ = make_blobs(n_samples=300, centers=2, cluster_std=0.60, random_state=0)
    X = np.vstack([X, np.random.uniform(low=-6, high=6, size=(50, 2))])
    X = StandardScaler().fit_transform(X)  # Standardize features
    return X

def plot_anomalies(X, y_pred, title):
    """
    Plot the data and the detected anomalies.
    """
    plt.figure(figsize=(10, 7))
    plt.title(title)
    plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='coolwarm', edgecolor='k')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.colorbar(label='Anomaly')
    plt.show()

def apply_isolation_forest(X):
    """
    Apply Isolation Forest for anomaly detection.
    """
    print("\n--- Isolation Forest ---")
    model = IsolationForest(contamination=0.1, random_state=42)
    y_pred = model.fit_predict(X)
    y_pred = np.where(y_pred == 1, 0, 1)  # Convert -1 to 1 for outliers
    plot_anomalies(X, y_pred, 'Isolation Forest Anomaly Detection')

def apply_lof(X):
    """
    Apply Local Outlier Factor (LOF) for anomaly detection.
    """
    print("\n--- Local Outlier Factor (LOF) ---")
    model = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
    y_pred = model.fit_predict(X)
    y_pred = np.where(y_pred == 1, 0, 1)  # Convert -1 to 1 for outliers
    plot_anomalies(X, y_pred, 'Local Outlier Factor (LOF) Anomaly Detection')

def apply_one_class_svm(X):
    """
    Apply One-Class SVM for anomaly detection.
    """
    print("\n--- One-Class SVM ---")
    model = OneClassSVM(gamma='auto', nu=0.1)
    model.fit(X)
    y_pred = model.predict(X)
    y_pred = np.where(y_pred == 1, 0, 1)  # Convert -1 to 1 for outliers
    plot_anomalies(X, y_pred, 'One-Class SVM Anomaly Detection')

def main():
    # Generate synthetic data
    X = generate_synthetic_data()

    # Apply each anomaly detection algorithm
    apply_isolation_forest(X)
    apply_lof(X)
    apply_one_class_svm(X)

if __name__ == "__main__":
    main()
```
